```python
# How many people of each race are represented in this dataset?
race_count = adult_data['race'].value_counts()
race_count
```

```python
# create subplots using grid-like format
fig = plt.figure(figsize=(14, 10))
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)

# set title
ax3.set_title('Count per Race')
ax1.set_title('Random')

# adjust height
# plt.subplots_adjust(hspace=0.4)

# set label
ax3.set_xlabel('Race')
ax3.set_ylabel('Count')

# plot data
ax3.plot(race_count, color='red', linewidth=3, marker='o', label='count')
ax1.plot(np.random.randn(50), c='green', linestyle='--')

# set legend
ax3.legend()
```
